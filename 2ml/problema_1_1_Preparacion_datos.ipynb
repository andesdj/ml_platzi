```python
financials= pd.read_csv('datasets/thenumbers.csv', encoding='utf-8')
financials.head()
```

```python
financials=financials[['movie_title', 'production_budget', 'worldwide_gross']]
```

```python
#Agregar la misma llave para hacer el JOIN  en el DF Numerico
```

```python
movies_num=pd.concat([movies_num, movies['movie_title']], axis=1)
```

```python
movies_num
```

```python
movies2=pd.merge(financials,movies_num, on='movie_title', how='left')
```

```python
#Detectar campos nulos
movies2.notnull()
```

```python
# Aplica columna por columna  sobre cada una de las columnas
movies2.notnull().apply(pd.Series.value_counts)
```

```python
(movies2 !=0).apply(pd.Series.value_counts)
```

```python
avaiable=((movies2!=0)& (movies2.notnull()))
avaiable.all(axis=1).value_counts()
```

```python
mask=avaiable['worldwide_gross']
movies2=movies2[mask]
```

```python

((movies2!=0)& (movies2.notnull())).worldwide_gross.value_counts()
```

```python
#SCikit Learn preprocesing para completar datos faltantes
from sklearn.impute import SimpleImputer
imputer= SimpleImputer(missing_values=np.nan, strategy='mean')
```

```python
movies2=movies2.drop('duration',axis=1)
movies2=movies2.drop('movie_title',axis=1)
```

```python
movies2
```

```python
values=imputer.fit_transform(movies2)
```

```python
x=pd.DataFrame(values)
x.columns = movies2.columns
x.index=movies2.index
x.head()
```

```python
x.to_csv('datasets/x.csv', index=False)
```
